## Problem Statment:

#### XYZ Company is offering career opportunity for its various project(s). The available job profiles are (a) Web Development and (b) Data Scientist. There are almost 500 candidates applied for the respective role. Your assignment is to build a Machine Learning model , which will help the company management to Identify the potential candidates whom they can offer job (to process further for interview). 

```python
import pandas as pd
from sklearn import linear_model
import math
```

```python
df=pd.read_excel("ml_sample_dataset.xlsx")
```

```python
df.describe()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=1,how='all')
```

### Transforming the Performances of 10,12,UG,PG

```python
df['Performance_10']=df['Performance_10'].astype(str)
df['Performance_10']=df['Performance_10'].str.split("/")
df['cgpa']=df['Performance_10'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_10_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_10_std']=df['cgpa']*10
```

### Predicting Performance_10 to replace null values

```python
df_ml= pd.DataFrame({'Performance_10_std':df['Performance_10_std']})
df_ml['Python (out of 3)']=df['Python (out of 3)']
df_ml['R Programming (out of 3)']=df['R Programming (out of 3)']
df_ml['Deep Learning (out of 3)']=df['Deep Learning (out of 3)']
df_ml['PHP (out of 3)']=df['PHP (out of 3)']
df_ml['MySQL (out of 3)']=df['MySQL (out of 3)']
df_ml['MongoDB (out of 3)']=df['MongoDB (out of 3)']
df_ml['HTML (out of 3)', ]=df['HTML (out of 3)']
df_ml['CSS (out of 3)']=df['CSS (out of 3)']
df_ml['JavaScript (out of 3)']=df['JavaScript (out of 3)']
df_ml['AJAX (out of 3)']=df['AJAX (out of 3)']
df_ml['Bootstrap (out of 3)']=df['Bootstrap (out of 3)']
df_ml['Node.js (out of 3)']=df['Node.js (out of 3)']
df_ml['ReactJS (out of 3)']=df['ReactJS (out of 3)']
df_ml=df_ml.dropna()

reg_model = linear_model.LinearRegression()
X_train=df_ml[['Python (out of 3)','R Programming (out of 3)','Deep Learning (out of 3)', 'PHP (out of 3)', 
              'MySQL (out of 3)', 'MongoDB (out of 3)', 'HTML (out of 3)', 'CSS (out of 3)', 'JavaScript (out of 3)',
              'AJAX (out of 3)', 'Bootstrap (out of 3)', 'Node.js (out of 3)', 'ReactJS (out of 3)']]
Y_train=df_ml[['Performance_10_std']]

reg_model.fit(X_train,Y_train)


for i in range (0,392):
    if(math.isnan(df.iloc[i, df.columns.get_loc('Performance_10_std')])):
        df1=df.iloc[i:i+1, 2:15]
        pred=reg_model.predict(df1[['Python (out of 3)','R Programming (out of 3)',
              'Deep Learning (out of 3)', 'PHP (out of 3)', 
              'MySQL (out of 3)', 'MongoDB (out of 3)', 'HTML (out of 3)', 'CSS (out of 3)', 'JavaScript (out of 3)',
              'AJAX (out of 3)', 'Bootstrap (out of 3)', 'Node.js (out of 3)', 'ReactJS (out of 3)']])
        df['Performance_10_std'][i]=float(pred)
```

### Transforming the Performances of 12

```python
df['Performance_12']=df['Performance_12'].astype(str)
df['Performance_12']=df['Performance_12'].str.split("/")
df['cgpa']=df['Performance_12'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_12_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_12_std']=df['cgpa']*10

```

### Predicting Performance_12 to replace null values

```python
df_ml= pd.DataFrame({'Performance_12_std':df['Performance_12_std']})
df_ml['Performance_10_std']=df['Performance_10_std']
df_ml=df_ml.dropna()

reg_model = linear_model.LinearRegression()
reg_model.fit(df_ml[['Performance_10_std']],df_ml[['Performance_12_std']])

for i in range (0,392):
    if(math.isnan(df.iloc[i, df.columns.get_loc('Performance_12_std')])):
        df1=df.iloc[i:i+1]
        pred=reg_model.predict(df1[['Performance_10_std']])
        df['Performance_12_std'][i]=float(pred)
```

### Transforming the Performances of UG

```python
df['Performance_UG']=df['Performance_UG'].astype(str)
df['Performance_UG']=df['Performance_UG'].str.split("/")
df['cgpa']=df['Performance_UG'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_UG_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_UG_std']=df['cgpa']*10
```

### Predicting Performance_UG to replace null values

```python
df_ml= pd.DataFrame({'Performance_UG_std':df['Performance_UG_std']})
df_ml['Performance_12_std']=df['Performance_12_std']
df_ml=df_ml.dropna()

reg_model = linear_model.LinearRegression()
reg_model.fit(df_ml[['Performance_12_std']],df_ml[['Performance_UG_std']])

for i in range (0,392):
    if(math.isnan(df.iloc[i, df.columns.get_loc('Performance_UG_std')])):
        df1=df.iloc[i:i+1]
        pred=reg_model.predict(df1[['Performance_12_std']])
        df['Performance_UG_std'][i]=float(pred)
```

### Transforming the Performances of PG

```python
df['Performance_PG']=df['Performance_PG'].astype(str)
df['Performance_PG']=df['Performance_PG'].str.split("/")
df['cgpa']=df['Performance_PG'].str.get(0)
df['cgpa']=df['cgpa'].astype('float')

#converting 10 CGPA to 100 percentage
df.loc[df['cgpa']>10,'Performance_PG_std']=df['cgpa']
df.loc[df['cgpa']<11,'Performance_PG_std']=df['cgpa']*10
```

### Predicting Performance_PG to replace null values

```python
df_ml= pd.DataFrame({'Performance_PG_std':df['Performance_PG_std']})
df_ml['Performance_UG_std']=df['Performance_UG_std']
df_ml=df_ml.dropna()

reg_model = linear_model.LinearRegression()
reg_model.fit(df_ml[['Performance_UG_std']],df_ml[['Performance_PG_std']])

for i in range (0,392):
    if(math.isnan(df.iloc[i, df.columns.get_loc('Performance_PG_std')])):
        df1=df.iloc[i:i+1]
        pred=reg_model.predict(df1[['Performance_UG_std']])
        df['Performance_PG_std'][i]=float(pred)
```

```python
#droping unnecessary columns
df=df.drop('cgpa',1)
df=df.drop('Performance_10',1)
df=df.drop('Performance_12',1)
df=df.drop('Performance_UG',1)
df=df.drop('Performance_PG',1)
```
